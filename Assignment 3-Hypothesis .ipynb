```python
n1=5
n2=5
x1bar=7.04314
x2bar=6.93726
sd1=0.294960
sd2=0.276628
alpha = 0.05
```

```python
SE = np.sqrt(sd1**2/n1+sd2**2/n2)
t = ((x1bar-x2bar)/SE)
print("calculate value of t test is: ",t)
```

```python
df=n1+n2-2
cv=stats.t.ppf(alpha,df)
print("critical value is: ",cv)
```

```python
pval=stats.t.cdf(t,df)
print("p value of right-sided test is: ",pval)
```

We can clearly see that there is a significant difference between p value = 0.7128156108298787 and t tests value = 0.5854739597300976

```python
#One Way ANOVA Example


# creating data frame 
d = {'Phillippines': pd.Series([271,29]),
     'Indonesia': pd.Series([267,33]),
     'Malta': pd.Series([269,31]),
     'India': pd.Series([280,20])}

df = pd.DataFrame(d)

# significance level
alpha = 0.05  
# Perfrom One Way ANOVA
outcome = stats.f_oneway(df['Phillippines'],df['Indonesia'],df['Malta'],df['India'])
# gettin p-value of test
p_value= outcome[1]


```

```python
#Null hypothesis mu_1 = mu_2 = mu_3
if p_value <= alpha:
    # we reject null hypothesis
    print ('Null hypothesis is unlikely to except')
else:
    # we reject alternative hypothesis
    print('Null hypothesis cannot be rejected')
```
